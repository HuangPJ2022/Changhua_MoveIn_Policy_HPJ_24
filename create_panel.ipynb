## compute growth rate

```python
growth_df = pd.DataFrame()

for month in range(1, 13):
    if month < 10:
        month_name = f'0{month}'
    else:
        month_name = month

    read_left_file_name = f'{folder_path}/{month}/{98}{month_name}_彰化鄉鎮市.xlsx'
    start_df = pd.read_excel(read_left_file_name).rename(columns = {"place" : "place", "num" : f'num_{98}{month_name}'})

    ## all county
    county_num = start_df[f'num_{98}{month_name}'].sum()
    county = "全縣"
    start_df.loc[len(start_df)] = [county, county_num]

    growth_df['place'] = start_df['place']
    left_num = start_df[f'num_{98}{month_name}']

    for year in range(99, 113):

        if year == 112 and month == 12:    
            break

        
        read_right_file_name = f'{folder_path}/{month}/{year}{month_name}_彰化鄉鎮市.xlsx'
        temp_df = pd.read_excel(read_right_file_name).rename(columns = {"place" : "place", "num" : f'num_{year}{month_name}'})

        ## all county
        county_num = temp_df[f'num_{year}{month_name}'].sum()

        temp_df.loc[len(temp_df)] = [county, county_num]

        right_num = temp_df[f'num_{year}{month_name}']
        
        growth_df[f'growth_{year}{month_name}'] = (right_num - left_num)/left_num

        left_num = right_num

growth_df.head(30)
```

## num

```python
num_df = pd.DataFrame()
total_num_df = pd.DataFrame()
total_size_df = pd.DataFrame()

num_df['place'] = growth_df['place']
total_num_df['place'] = growth_df['place']
total_size_df['place'] = growth_df['place']

for month in range(1, 13):
    if month < 10:
        month_name = f'0{month}'
    else:
        month_name = month
    
    ## all county
    county = "全縣"
    for year in range(99, 113):

        if year == 112 and month == 12:    
            break
        
        read_file_name = f'{folder_path}/{month}/{year}{month_name}_彰化鄉鎮市.xlsx'
        temp_df = pd.read_excel(read_file_name).rename(columns = {"place" : "place", "num" : f'num_{year}{month_name}'})

        read_total_file_name = f'{folder_path_2}/{month}/{year}{month_name}_彰化鄉鎮市.xlsx'
        temp_df_total = pd.read_excel(read_total_file_name).rename(columns = {"place" : "place", "num" : f'total_num_{year}{month_name}'})
        total_num_li = temp_df_total[f'total_num_{year}{month_name}'].to_list()

        for t in range(len(total_num_li)):
            total_num_li[t] = total_num_li[t].replace(',','')
            total_num_li[t] = int(total_num_li[t])

        temp_df_total[f'total_num_{year}{month_name}'] = total_num_li

        ## all county
        county_growth = temp_df[f'num_{year}{month_name}'].sum()
        county_num = temp_df_total[f'total_num_{year}{month_name}'].sum()

        temp_df.loc[len(temp_df)] = [county, county_growth]
        temp_df_total.loc[len(temp_df_total)] = [county, county_num]

        
        order_li = [0]*27
        to_rank_order = temp_df_total[temp_df_total.columns[1]].to_list()
        for size in range(27):
            order = to_rank_order.index(max(to_rank_order))
            order_li[order] = (27 - size)
            to_rank_order[order] = 0
    
        total_size_df[f'city_size_{year}{month_name}'] = order_li

        num_df = pd.merge(left = num_df, right = temp_df, how = 'left')
        total_num_df = pd.merge(left = total_num_df, right = temp_df_total, how = 'left')
```

```python
df2 = pd.DataFrame(columns = ['year','month'])

y_li = []
m_li = []

for t in range(27):
    for i in range(99,113):
        for j in range(1, 13):
            if i == 112 and j == 12:
                break

            y_li.append(i)
            m_li.append(j)

df2['year'] = y_li
df2['month'] = m_li

place_li = []
id_li = []
growth_li = []
num_li = []
total_num_li = []
size_order_li = []
ym_li = []

for i in range(27):
    growth_row = growth_df.iloc[i][1:].to_list()
    num_row = num_df.iloc[i][1:].to_list()
    total_row = total_num_df.iloc[i][1:].to_list()
    size_order = total_size_df.iloc[i][1:].to_list()
    place = num_df.iloc[i][0]

    ym = 0
    for j in range(167):
        ym += 1
        ym_li.append(ym)
        place_li.append(place)
        id_li.append(i+1)
        growth_li.append(growth_row[j])
        num_li.append(num_row[j])
        total_num_li.append(total_row[j])
        size_order_li.append(size_order[j])


df2['ym'] = ym_li
df2['place'] = place_li
df2['place_id'] = id_li
df2['growth'] = growth_li
df2['num'] = num_li
df2['total_num'] = total_num_li
df2['city_size'] = size_order_li

df2['policy'] = [0]*4509
df2['after_policy'] = [0]*4509

for i in range(4509):
    if int(df2['year'].iloc[i]) == 111:
        if int(df2['month'].iloc[i]) > 2 and int(df2['month'].iloc[i]) < 7:
            df2['after_policy'][i] = np.nan
            df2['policy'][i] = 1

    if int(df2['year'].iloc[i]) == 112:
            df2['policy'][i] = np.nan
            df2['after_policy'][i] = 1
```
